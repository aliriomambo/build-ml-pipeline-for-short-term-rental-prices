```python
import wandb
import pandas as pd
import ydata_profiling
```

```python
run = wandb.init(project="nyc_airbnb", group="eda", save_code=True)
local_path = wandb.use_artifact("sample.csv:latest").file()
df = pd.read_csv(local_path)
```

# Generate EDA Report for Analysis

```python
profile = ydata_profiling.ProfileReport(df)
profile
```

# Fixing Dataset values

```python
# Fill missing values in 'reviews_per_month' with 0
df['reviews_per_month'].fillna(0, inplace=True)

# Convert 'last_review' to datetime, coercing errors will turn unparseable dates into NaT
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')

# Filter the DataFrame to keep only rows where the price is between $10 and $350
df = df.query('10 <= price <= 350').copy()

# Generate a new profile report on the cleaned data
clean_profile = ydata_profiling.ProfileReport(df, title="Cleaned Pandas Profiling Report")
clean_profile
```

# Save the dataset

```python
run.finish()
```
